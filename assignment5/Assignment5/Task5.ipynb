```python
users['last_session_creation_time']=pd.to_datetime(users['last_session_creation_time'],unit='s')
users['creation_time']=pd.to_datetime(users['creation_time'])

users
```

```python
user_Engagement=pd.read_csv("/content/sample_data/takehome_user_engagement.csv")
user_Engagement
```

```python
user_Engagement.info()
```

```python
user_Engagement["time_stamp"]=pd.to_datetime(user_Engagement["time_stamp"])
```

```python
df1=user_Engagement.copy()
df1
```

```python
Pivot=pd.pivot_table(data=user_Engagement,index="user_id",values='time_stamp')
Pivot
```

```python
df1['visits_in_7_days']=df1.groupby('user_id',as_index=False,group_keys=False).apply(lambda x : x.rolling('7D', on="time_stamp")['user_id'].count())
df1.head()
```

```python
df1.describe().T
```

```python
df1[df1.visits_in_7_days>=3.0]
df1
```

```python
adopted=df1.groupby('user_id')['visits_in_7_days'].max().reset_index()
adopted.head()
```

```python
adopted['is_adopted_user']=adopted['visits_in_7_days'].apply(lambda x: 1 if x>=3 else 0)
adopted.head()
```

```python
adopted.is_adopted_user.value_counts()
```

```python
adopted.drop('visits_in_7_days', axis = 1, inplace=True)
adopted.head()
```

```python
adopted.rename(columns={'user_id':'object_id'}, inplace=True)
adopted
```

```python
adopted.set_index('object_id',inplace=True)
adopted
```

```python
users1=users.copy()
users1
```

```python
new_users_df=pd.merge(users1,adopted,on='object_id',how='left')
new_users_df
```

```python
new_users_df.shape
```

```python
new_users_df['last_session_creation_time'].fillna(0,inplace=True)
new_users_df['is_adopted_user'].fillna(0,inplace=True)
```

```python
new_users_df.describe().T
```

```python
new_users_df['email_domain'] = new_users_df.email.apply(lambda x: x.split('@')[1])
```

```python
new_users_df['email_domain'].value_counts()
```

There are too many email domains and most of them seem fake domains so it's good to drop the column entirely. We can also drop the name and object_id columns.

And for the invited_by_user_id let's convert the NULL values to 0 because the column has a Non Null value only if the creation_source was a GUEST_INVITE or a ORG_INVITE anyways.

For the creation_time column let's add a column which calculates how old the account is, i.e. the number of days since the account was created.

And since last_session_creation_time can be removed as well because it was in a sense used to create the adopted_user column.

```python
new_users_df.drop(['object_id', 'name', 'email', 'email_domain'], axis = 1, inplace = True)
```

```python
new_users_df.invited_by_user_id.fillna(0, inplace=True)
```

```python
new_users_df['days_since_creation'] = (df1.time_stamp.max() - new_users_df.creation_time).dt.days
```

```python
new_users_df.drop(['creation_time', 'last_session_creation_time'], axis = 1, inplace = True)
```

```python
new_users_df = pd.get_dummies(new_users_df, columns=['creation_source'])
```
